```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Imort Dataset
df=pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv',index_col=0)
```

```python
#View Dataset
df
```

```python
#Data is not considered na, because values other than na are not considered missing
df.isna().sum().sum()
```

```python
#determining if there are any missing values
df['weight'].isnull()
```

```python
#Dropping columns
df2=df.drop(columns=['admission_type_id','discharge_disposition_id','admission_source_id','payer_code','medical_specialty'],axis=1)
```

```python
#Viewing new dataset
df2
```

```python
#Dropping more columns
df3=df2.drop(columns=['metformin','repaglinide','nateglinide','chlorpropamide','glimepiride','acetohexamide','glipizide-metformin','glyburide','tolbutamide','pioglitazone','rosiglitazone','acarbose','miglitol','troglitazone','tolazamide','examide','citoglipton','insulin','glyburide-metformin','glipizide-metformin','glimepiride-pioglitazone','metformin-rosiglitazone','metformin-pioglitazone','max_glu_serum','weight'],axis=1)
```

```python
#Viewing Dataset
df3
```

```python
#identifying duplicates
df3.sort_values('patient_nbr',inplace=True)
```

```python
#dropping duplicates
df3.drop_duplicates(subset='patient_nbr',
                   keep=False,inplace=True)
```

```python
#Viewing new dataset
df3
```

```python
#Higher A1C level is, poorer your blood sugar
```

```python
#Dropping more columns
df4=df3.drop(columns=['num_procedures','num_medications','number_outpatient','num_lab_procedures','time_in_hospital','num_medications','number_outpatient','number_inpatient'],axis=1)
```

```python
#Viewing dataset
df4
```

```python
#Viewing info
df4.info()
```

```python
#viewing graph
df4.groupby('gender').age.hist()
```

```python
#plotting based on age and gender
sns.countplot(x='age',hue='gender',data=df4)
```

```python
#plotting based off of age and race
sns.countplot(x='age',hue='race',data=df4)
```

```python
#Viewing sums of each age category
df4['age'].value_counts(ascending=True)
```

```python
#viewing sums of each race category
df4['race'].value_counts(ascending=True)
```

```python
#Male Count
df5=df4[df4['gender']=='Male']
sns.countplot(x='age',hue='race',data=df5)
```

```python
#Female Count
df6=df4[df4['gender']=='Female']
sns.countplot(x='age',hue='race',data=df6)
```
